```python
# 3. Compare and contrast mutable and immutable objects in Python With examples.
```

```python
"""
objects are classified as either mutable or immutable based on 
whether their values can be changed after creation."""

# Example of Mutable Object
my_list = [1, 2, 3]
print(id(my_list))  # Prints the memory address of my_list
```

```python
"Immutable Objects
'''Definition:** Objects whose values cannot be changed after creation.
Examples:** Integers, floats, strings, tuples, booleans.
Behavior:** When you modify an immutable object, a new object is created with the updated value.
Pros: Safer for data integrity, especially in concurrent programming.
Cons: Can be less memory efficient when creating many new objects'''

# Example of Immutable Object

my_string = "Hello"
print(id(my_string))  # Prints the memory address of my_string

my_string += " World"
print(my_string)
print(id(my_string))
```

```python
# 4. Discuss the different types of operators in Python and provide examples of how they are used.
```

```python
# Arithmetic Operators
"""Arithmetic operators are used to perform mathematical operations such as 
addition, subtraction, multiplication, etc."""

# '+': Adds two operands.
x = 5
y = 3
print(x + y)

# '-': Subtracts the second operand from the first.
print(x - y) 

# '*': Multiplies two operands.
print(x * y)

# '/': Divides the first operand by the second.
print(x / y)

# '%' : Returns the remainder when the first operand is divided by the second (Modulus).
print(x % y)

# '**': Raises the first operand to the power of the second (Exponentiation).
print(x ** y)

# '//' : Performs division and returns the integer part of the quotient (Floor Division).
print(x // y)


"Assignment Operators
"""Assignment operators are used to assign values to variables."""

# '=': Assigns a value to a variable.
x = 5

# '+=' : Adds and assigns the result.
x += 3  # Equivalent to x = x + 3
print(x)

# '-=' : Subtracts and assigns the result.
x -= 2  # Equivalent to x = x - 2
print(x)

# '*=': Multiplies and assigns the result.
x *= 2  # Equivalent to x = x * 2
print(x)

# '/=': Divides and assigns the result.
x /= 4  # Equivalent to x = x / 4
print(x)

# '%=' : Takes modulus and assigns the result.
x %= 3  # Equivalent to x = x % 3
print(x)

# '**=': Raises to power and assigns the result.
x = 2
x **= 3  # Equivalent to x = x ** 3
print(x)

# '//=': Performs floor division and assigns the result.
x = 9
x //= 2  # Equivalent to x = x // 2
print(x) 

"Comparison Operators
'''Comparison operators compare two values and return a Boolean result (True or False).'''

# '==': Checks if two values are equal.
print(x == 4)  # Output: True

# '!=' (Not equal to): Checks if two values are not equal.
print(x != 4)  # Output: False

# '>': Checks if the first value is greater than the second.
print(x > 2)  # Output: True

# '<': Checks if the first value is less than the second.
print(x < 2)  # Output: False

# '>=': Checks if the first value is greater than or equal to the second.
print(x >= 4)  # Output: True

# '<=': Checks if the first value is less than or equal to the second.
print(x <= 4)  # Output: True

"Logical Operators
'''Logical operators are used to combine conditional statements.'''

# 'and': Returns 'True' if both statements are true.
print(x > 2 and x < 5)  # Output: True

# 'or': Returns 'True' if at least one of the statements is true.
print(x > 2 or x < 2)  # Output: True

# 'not': Reverses the result, returns 'False' if the result is true.
print(not(x > 2))  # Output: False

"Bitwise Operators
'''Bitwise operators are used to perform operations on binary numbers.'''

# '&' : Sets each bit to 1 if both bits are 1.
a = 5  # 0101 in binary
b = 3  # 0011 in binary
print(a & b)  # Output: 1 (0001 in binary)

# '|' (OR): Sets each bit to 1 if one of the bits is 1.
print(a | b)  # Output: 7 (0111 in binary)

# '^' (XOR): Sets each bit to 1 if only one of the bits is 1.
print(a ^ b)  # Output: 6 (0110 in binary)

# '~' (NOT): Inverts all the bits.
print(~a)  # Output: -6 (inverts 0101 to 1010, which is -6 in two's complement)

# '<<' (Zero fill left shift): Shifts left by pushing zeros in from the right.
print(a << 1)  # Output: 10 (1010 in binary)

# '>>' (Signed right shift): Shifts right by pushing copies of the leftmost bit in from the left.
print(a >> 1)  # Output: 2 (0010 in binary)

"Membership Operators
'''Membership operators are used to test if a sequence is presented in an object.'''

# 'in': Returns 'True' if a sequence with the specified value is present in the object.
lst = [1, 2, 3, 4]
print(3 in lst)  # Output: True

# 'not in': Returns 'True* if a sequence with the specified value is not present in the object.
print(5 not in lst)  # Output: True

" Identity Operators
'''Identity operators are used to compare the memory locations of two objects.'''

# 'is': Returns 'True' if both variables are the same object.
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x
print(x is z)  # Output: True
print(x is y)  # Output: False

# 'is not': Returns 'True* if both variables are not the same object.
print(x is not y)  # Output: True

```

```python
# 5. Explain the concept of type casting in Python with examples.
```

```python
'''This can be done either explicitly by the programmer or implicitly by the Python interpreter. 
Explicit type casting is also known as explicit type conversion, and it is done using built-in functions.'''

# Explicit Type Casting
# Explicit type casting is performed using Python's built-in functions. Here are some common functions for type casting:

# int(): Converts a value to an integer.
# float(): Converts a value to a float.
# str(): Converts a value to a string.
# list(): Converts a value to a list.
# tuple(): Converts a value to a tuple.
# set(): Converts a value to a set.
# dict(): Converts a value to a dictionary (from a sequence of key-value pairs).

"Integer to Float
x = 10
y = float(x)
print(y)  # Output: 10.0

"Float to Integer
x = 10.5
y = int(x)
print(y)  # Output: 10

"Integer to String
x = 10
y = str(x)
print(y)  # Output: '10'

"String to Integer
x = "10"
y = int(x)
print(y)  # Output: 10


"String to Float
x = "10.5"
y = float(x)
print(y)  # Output: 10.5

"List to Tuple
x = [1, 2, 3]
y = tuple(x)
print(y)  # Output: (1, 2, 3)

"Tuple to List
x = (1, 2, 3)
y = list(x)
print(y)  # Output: [1, 2, 3]

"List to Set
x = [1, 2, 2, 3, 4]
y = set(x)
print(y)  # Output: {1, 2, 3, 4}

"List of Tuples to Dictionary
x = [("a", 1), ("b", 2)]
y = dict(x)
print(y)  # Output: {'a': 1, 'b': 2}



"Implicit Type Casting
'''Implicit type casting is done automatically by the Python interpreter without the need for the programmer 
to specify the type conversion. This usually happens in expressions where different data types are mixed.'''


"Integer to Float
x = 10
y = 2.5
z = x + y
print(z)  # Output: 12.5
print(type(z))  # Output: <class 'float'>

"Integer to String (Concatenation)
x = 10
y = " apples"
z = str(x) + y
print(z)  # Output: '10 apples'

#Inside: In the first example, Python automatically converts the integer x to a float to perform the addition with the float y. 
#In the second example, Python converts the integer x to a string before concatenating it with the string y.
```

```python
# 7. Describe the different types of loops in Python and their use cases with examples.
```

```python
"""Python provides several types of loops that allow you to execute a block of code multiple times. 
The primary loop constructs in Python are for and while loops. 
Additionally, break and continue statements can be used to control the flow within these loops."""

" 'for' Loop
'''A 'for' loop is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string). 
This is the preferred loop when you know the number of iterations beforehand or 
when you want to iterate over elements in a sequence.'''

"Syntax:

for element in sequence:
    # code block

"Examples:

"Iterating Over a List:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output:
# apple
# banana
# cherry

"Iterating Over a Range of Numbers:

for i in range(5):
    print(i)
# Output:
# 0
# 1
# 2
# 3
# 4

"Iterating Over a Dictionary:

student = {"name": "John", "age": 20, "grade": "A"}
for key, value in student.items():
    print(key, value)

# Output:
# name John
# age 20
# grade A

"while Loop
'''A while loop repeatedly executes a block of code as long as the given condition is true. 
This loop is useful when the number of iterations is not known beforehand 
and the loop should continue until a certain condition is met.'''

"Examples:

"Basic 'while' Loop:
i = 0
while i < 5:
    print(i)
    i += 1
# Output:
# 0
# 1
# 2
# 3
# 4

"Infinite Loop with Break Condition:

i = 0
while True:
    print(i)
    i += 1
    if i >= 5:
        break
# Output:
# 0
# 1
# 2
# 3
# 4

"Controlling Loops with 'break' and 'continue'
"break Statement:
# The break statement is used to exit a loop prematurely when a certain condition is met.

"Example:
for i in range(10):
    if i == 5:
        break
    print(i)

# Output:
# 0
# 1
# 2
# 3
# 4

" 'continue' Statement:
'''The continue statement is used to skip the current iteration and proceed to the next iteration of the loop.'''

"Example:
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)

# Output:
# 1
# 3
# 5
# 7
# 9
```
